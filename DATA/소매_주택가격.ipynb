```python
import pandas as pd

df = pd.read_excel('종로가게데이터.xlsx')
```

```python
df['인허가일자'] = df['인허가일자'].astype('str')
df['폐업일자'] = df['폐업일자'].astype('str')
pd.to_datetime(df['인허가일자'])
pd.to_datetime(df['폐업일자'])
```

```python
df
```

```python
df['개업년도'] = ''
df['폐업년도'] = ''
```

```python
df
```

```python
for i in range(len(df)) :
    if int(df['인허가일자'][i][:4]) < 2011 :
        df['개업년도'][i] = '2011'
    else :
        df['개업년도'][i] = df['인허가일자'][i][:4]
    if df['폐업일자'][i] == 'NaT' or int(df['폐업일자'][i][:4]) > 2019 :
        df['폐업년도'][i] = '2019'
    else :
        df['폐업년도'][i] = df['폐업일자'][i][:4]
```

```python
retail_data = pd.read_excel('사업체현황최종(도매소매).xlsx')
house_data = pd.read_excel('주택가격최종(최종).xlsx')
```
