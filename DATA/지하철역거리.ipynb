```python
cal=pd.DataFrame()
cal.append({"name":"","diff":"","ranks":""},ignore_index=True)
```

```python
df = pd.DataFrame()
```

```python
for i in range(0,len(data)):
    for j in range(0,len(station)):
        cal.loc[j,'name']=station['역이름'][j]
        cal.loc[j, 'diff'] = haversine((data['위도'][i], data['경도'][i]), (station['y'][j], station['x'][j]), unit='m')
    cal['ranks']=cal['diff'].rank(ascending=True)
    df.loc[i,'rank1']=cal.loc[cal['ranks']==1,'name'].iloc[0]
    df.loc[i,'rank1_diff']=cal.loc[cal['ranks']==1,'diff'].iloc[0]
```
