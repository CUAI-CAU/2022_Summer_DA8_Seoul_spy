```python
gyolist=df[1].tolist()
```

```python
 # 주소 -> 위도 경도 변환
import requests
from urllib import parse

def YLogin(addr):
    headers = {
        "Content-Type":"application/x-www-form-urlencoded"
    }
    yData = "refine_ty=8&protocol_=" + parse.quote(addr)

    yUrl = "https://address.dawul.co.kr/input_pro.php"
    session = requests.Session()

    try:
        resp = session.post(url = yUrl, headers = headers, data = yData)
        data = resp.text.split("|")
        return data[-2] + " " + data[-1]
    except requests.exceptions.Timeout:
        return "Timeout"
    except requests.exceptions.HTTPError:
        return "Http error"
    except requests.exceptions.ConnectionError:
        return "Connection error"

if __name__ == "__main__":
    latlist=[]
    for i in gyolist:
      latlist.append(YLogin(i))
```
