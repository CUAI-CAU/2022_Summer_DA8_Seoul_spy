```python
c=[]
for i in range(len(a)):
    if b[i]<pd.Timestamp(2011,1,1) or a[i]>pd.Timestamp(2019,12,31):
        c.append(None)
    elif a[i]<=pd.Timestamp(2011,1,1)  and b[i]>=pd.Timestamp(2019,12,31):
        c.append(pd.Timestamp(2019,12,31)-pd.Timestamp(2011,1,1))
    elif a[i]>pd.Timestamp(2011,1,1)  and  b[i]>=pd.Timestamp(2019,12,31):
        c.append(pd.Timestamp(2019,12,31)-a[i])
    elif a[i]>=pd.Timestamp(2011,1,1)  and  b[i]<pd.Timestamp(2019,12,31) :
        c.append(b[i]-a[i])
    elif a[i]<pd.Timestamp(2011,1,1)  and  b[i]>pd.Timestamp(2019,12,31):
        c.append(b[i]-pd.Timestamp(2011,1,1))
    else:
        c.append(pd.Timestamp(2019,12,31)-a[i])
           
data['영업일']=c
```
