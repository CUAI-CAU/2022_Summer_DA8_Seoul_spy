```python
import pandas as pd
data=pd.read_excel('C:/Users/user/Downloads/데이터통합.xlsx')
data
```

```python
data1=data.loc[:,['동','상주인구 증가율','프랜차이즈점 개수','평균영업기간(년)','창폐업수','유동인구 증가율','매출액증가율']]
```

```python
data1.isna().sum()             #항동 2022년 5월 
```

```python
data1.dropna(axis=0,inplace=True)
```

```python
창폐업_횟수=[]
for row in data1['창폐업수']:
    if row >= data1['창폐업수'].mean()+2*data1['창폐업수'].std():
        창폐업_횟수.append(2)
    elif row >=data1['창폐업수'].mean()+data1['창폐업수'].std() and row <data1['창폐업수'].mean()+2*data1['창폐업수'].std() :
        창폐업_횟수.append(1)
    elif row >=data1['창폐업수'].mean() and  row <data1['창폐업수'].mean()+data1['창폐업수'].std() :
        창폐업_횟수.append(-1)
    else:
        창폐업_횟수.append(-2)
data1['창폐업 횟수'] = 창폐업_횟수    
```

```python
프랜차이즈_업체_수=[]
for row in data1['프랜차이즈점 개수']:
    if row >= data1['프랜차이즈점 개수'].mean()+2*data1['프랜차이즈점 개수'].std():
        프랜차이즈_업체_수.append(2)
    elif row >=data1['프랜차이즈점 개수'].mean()+data1['프랜차이즈점 개수'].std() and row <data1['프랜차이즈점 개수'].mean()+2*data1['프랜차이즈점 개수'].std() :
        프랜차이즈_업체_수.append(1)
    elif row >=data1['프랜차이즈점 개수'].mean() and  row <data1['프랜차이즈점 개수'].mean()+data1['프랜차이즈점 개수'].std() :
        프랜차이즈_업체_수.append(-1)
    else:
        프랜차이즈_업체_수.append(-2)
        
data1['프랜차이즈 업체 수'] = 프랜차이즈_업체_수       
```

```python
유동인구_증가율=[]
for row in data1['유동인구 증가율']:
    if row >= data1['유동인구 증가율'].mean()+2*data1['유동인구 증가율'].std():
        유동인구_증가율.append(2)
    elif row >=data1['유동인구 증가율'].mean()+data1['유동인구 증가율'].std() and row <data1['유동인구 증가율'].mean()+2*data1['유동인구 증가율'].std() :
        유동인구_증가율.append(1)
    elif row >=data1['유동인구 증가율'].mean() and  row <data1['유동인구 증가율'].mean()+data1['유동인구 증가율'].std() :
        유동인구_증가율.append(-1)
    else:
        유동인구_증가율.append(-2)
        
data1['유동인구_증가율'] = 유동인구_증가율      
```

```python
매출액_증가율=[]
for row in data1['매출액증가율']:
    if row >= data1['매출액증가율'].mean()+2*data1['매출액증가율'].std():
        매출액_증가율.append(2)
    elif row >=data1['매출액증가율'].mean()+data1['매출액증가율'].std() and row <data1['매출액증가율'].mean()+2*data1['매출액증가율'].std() :
        매출액_증가율.append(1)
    elif row >=data1['매출액증가율'].mean() and  row <data1['매출액증가율'].mean()+data1['매출액증가율'].std() :
        매출액_증가율.append(-1)
    else:
        매출액_증가율.append(-2)
        
data1['매출액 증가율'] = 매출액_증가율      
```

```python
#역변환
상주인구_증가율=[]
for row in data1['상주인구 증가율']:
    if row <= data1['상주인구 증가율'].mean()-2*data1['상주인구 증가율'].std():
        상주인구_증가율.append(2)
    elif row <=data1['상주인구 증가율'].mean()-data1['상주인구 증가율'].std() and row >data1['상주인구 증가율'].mean()-2*data1['상주인구 증가율'].std() :
        상주인구_증가율.append(1)
    elif row <=data1['상주인구 증가율'].mean() and  row >data1['상주인구 증가율'].mean()-data1['상주인구 증가율'].std() :
        상주인구_증가율.append(-1)
    else:
        상주인구_증가율.append(-2)
        
data1['상주인구_증가율'] = 상주인구_증가율       
```

```python
#역변환
평균_영업기간=[]
for row in data1['평균영업기간(년)']:
    if row <= data1['평균영업기간(년)'].mean()-2*data1['평균영업기간(년)'].std():
        평균_영업기간.append(2)
    elif row <=data1['평균영업기간(년)'].mean()-data1['평균영업기간(년)'].std() and row >data1['평균영업기간(년)'].mean()-2*data1['평균영업기간(년)'].std() :
        평균_영업기간.append(1)
    elif row <=data1['평균영업기간(년)'].mean() and  row >data1['평균영업기간(년)'].mean()-data1['평균영업기간(년)'].std() :
        평균_영업기간.append(-1)
    else:
        평균_영업기간.append(-2)
        
data1['평균 영업기간'] = 평균_영업기간  
```
